```python
from random import randint

numero_secreto = randint(1, 5)

while True:
  numero_escolhido = int(input(f' Escolha um número de 1 a 5: '))
  if numero_escolhido == numero_secreto:
      print(f'Parabéns você adivinhou o número secreto {numero_secreto} !')
      break
  else:
      print('Tente outra vez!')
```

```python
from random import randint

# criando o cabeçalho
frase_abertura = " Seja bem vindo ao jogo de adivinhação! "
print('-' * len(frase_abertura))
print(frase_abertura)
print('-' * len(frase_abertura))


# setando a variável com um número aleatório na faixa de 1 à 5.
numero_secreto = randint(1,5)
tentativa = 0
numero_escolhido = 0
acertou = False

while numero_secreto != numero_escolhido and tentativa < 3:
  if numero_secreto != numero_escolhido:
    numero_escolhido = int(input('informe um número de 1 a 5: '))
  if numero_secreto == numero_escolhido:
    acertou = True
  else:
    tentativa += 1

if acertou:
  print(f'Parabéns! Você descobriu que o número secreto era o {numero_secreto}!')
else:
  print(f'Desculpe, não foi desta vez! O número sercreto era {numero_secreto}')
```
