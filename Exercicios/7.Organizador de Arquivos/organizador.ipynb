```python
# Criando uma lista APENAS com os arquivos existentes, usando list comprehensions
# Para não termos problemas qto a letras maiúsculas/minúsculas, usar o lower() para tornar todas minúsculas
lista_arquivos = [arquivo.lower() for arquivo in os.listdir() if os.path.isfile(arquivo)]
lista_arquivos
```

```python
# Criando uma lista apenas com os tipos dos arquivos existentes
# Usando set para converter para conjunto e manter apenas 1 tipo de cada OU usando {} em vez de []
# lista_tipo = set([tipo.split(".")[-1] for tipo in lista_arquivos])
# Relembrando: conjunto são desornados/ não possuem índices.

lista_tipo = {tipo.split(".")[-1] for tipo in lista_arquivos}
lista_tipo
```

```python
# Criando as pastas de acordo com os arquivos existentes: o nome das pastas será o tipo do arquivo
# Antes: Verificando se as pastas existentes

for tipo in lista_tipo:
    if os.path.exists(tipo):
        pass
    else:
        os.mkdir(tipo)
```

```python
# Movendo os arquivos para as suas pastas
# de (caminho ABSOLUTO do arquivo!!)

for arquivo in lista_arquivos:
    pasta_destino = arquivo.split(".")[-1]
    de = os.path.join(os.getcwd(), arquivo)
    para = os.path.join(os.getcwd(), pasta_destino, arquivo)
    if os. path.exists(de):
        os.replace(de, para)
```
