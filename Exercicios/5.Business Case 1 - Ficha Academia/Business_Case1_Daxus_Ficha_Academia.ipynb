```python
nome = input("Digite o nome do aluno(a): ")
idade = input("Digite a idade do aluno(a): ")
peso = input("Digite o peso do aluno(a) (kg): ")
altura = input("Digite a altura do aluno(a) (m): ")
```

```python
# Calculando o IMC
# peso e altura inicialmente são STR/ tudo que vem do input é STR

imc = float(peso) / (float(altura) ** 2)

status = "Dentro do peso"

if imc > 25:
    status = "Sobrepeso"
elif imc < 18.5:
    status = "Abaixo do peso"

print(f"Seu IMC é: {imc:.2f}")
print(f"Você está: {status}")
```

```python
dados_aluno = {
    "nome": nome,
    "idade": idade,
    "peso": peso,
    "altura": altura,
    "imc": round(imc,2),
    "status": status
}
```
