- str(): para texto.

```python
nome = "Lucas"
numero_1 = 10
numero_2 = 28.8
print(type(nome))
print(type(numero_1))
print(type(numero_2))
```

### Listas:
- Conjunto ordenados de elementos.
-  Mutáveis.
- Heterogêneas.
- Possuem índice.

```python
lista = ['Lucas', 'Gabriel', 10, 250.80, False]
print(type(lista))
print(lista)
print(lista[2])
print(lista[-1])
```

```python
# adicionando um novo elemento na lista
lista.append('curso python')
print(lista)
```

```python
# removendo elemento na lista
lista.remove(False)
print(lista)
```

```python
# incluindo lista em lista
cores = ['azul','vermelho','verde']
lista.append(cores)
print(lista)
```

```python
# acessando o elemento verde
lista[6][2]
```

### Dicionários:
- Conjunto de elementos formado por chave: valor.
- Mutáveis.
- Heterogêneas.

```python
usuario = {
    'nome': 'Lucas',
    'Acesso': 'Junior',
    'nivel_acesso': 2
}
print(type(usuario))
print(usuario)

print(usuario['nome'])
print(usuario['Acesso'])
print(usuario['nivel_acesso'])
```

```python
# listando as chaves de um dicionário
print(usuario.keys())
```

```python
# listando os valores de um dicionário
print(usuario.values())
```

```python
# adicionando mais elementos a um dicionário - update()
usuario.update({'utimo_acesso':'2026-10-2'})
print(usuario)
```

```python
# removendo elementos de um dicionário
usuario.pop('nivel_acesso')
print(usuario)
```

```python
# alterando algum elemento de um dicionário
usuario['nome']= 'Gabriel'
print(usuario)
```

### Estrutura de Repetição - For

```python
nomes = ['Taíta','Lucas','Gabril','Lincon','Luna']

for nome in nomes:
  print(nome)
```

### range()

```python
i = 1
for numero in range(10):
  print(f"NÚmero {i}: {numero}")
  i = i+1
```

```python
# criando uma lista de numeros
numero = []
for n in range(10):
  numero.append(n)
print(numero)
```

```python
# Comando for com dicionários
usuario = {
    'nome': 'Lucas',
    'Acesso': 'Junior',
    'nivel_acesso': 2}

for chave in usuario.keys():
  print(chave)

print('\n')

for valor in usuario.values():
  print(valor)

print('\n')

for chave, valor in usuario.items():
  print(f'{chave}: {valor}')
```

### Estrutura condicional - IF

```python
idade = 5
if idade >= 18:
  print("Maior de idade, pode dirigir!!")
else:
  print("Menor de idade, não pode dirigir")
```

```python
idades =[10, 15, 35, 80, 25, 54, 10, 23, 18]
for idade in idades:
  print(f'Idade: {idade}')
  if idade >= 21:
    print("Pode consumir bebidaa alcóolica.")
  else:
    print("Não pode consumir bebida alcoolica.")
```

### Funções:
- Deixam o código mais limpo e organizado.
- Possível reaproveitamento de código.
- Palavra reservada: def.

```python
def mostrar_nome():
  print("Lucas")

mostrar_nome()
```

```python
def somar_numeros(num1, num2):
  print(f"A soma dos números é {num1 + num2}")

somar_numeros(10,200)
```

```python
def multiplica_numeros(num1, num2):
  return num1 * num2

multiplica_numeros(10,200)
```
